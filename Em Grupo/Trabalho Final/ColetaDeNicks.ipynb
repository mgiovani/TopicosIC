```python
import pandas as pd
from lxml import html
import requests
```

```python
lol = pd.DataFrame( {'nick' : [],
                     'classificacao' : []})

url = 'http://br.op.gg/ranking/ladder/page=1'
    
request = requests.get(url)
print(request, url)
html2 = html.fromstring(request.content)

teste = str(html2.xpath('//a[@class="ranking-highest__name"]/text()')).replace('[', '').replace(']', '').replace('\'', '')
teste2 = teste.split(',')

tester = str(html2.xpath('//div[@class="ranking-highest__rank"]/text()')).replace('[', '').replace(']', '').replace('\'', '')
tester2 = tester.split(',')

tst = pd.DataFrame(teste2, columns=['nick'])
tst2 = pd.DataFrame(tester2, columns=['classificacao'])
ts = pd.concat([tst, tst2], axis=1, join_axes=[tst.index])
result = [lol, ts]
lol = pd.concat(result)
```

```python
#13812 pginas no total
for page in range(1, 14550):
    url = 'http://br.op.gg/ranking/ladder/page='+str(page)
    
    request = requests.get(url)
    print(request, url)
    html2 = html.fromstring(request.content)

    teste = str(html2.xpath('//td[@class="ranking-table__cell ranking-table__cell--summoner"]/a/span/text()')).replace('[', '').replace(']', '').replace('\'', '')
    teste2 = teste.split(',')

    tester = str(html2.xpath('//td[@class="ranking-table__cell ranking-table__cell--rank"]/text()')).replace('[', '').replace(']', '').replace('\'', '')
    tester2 = tester.split(',')

    tst = pd.DataFrame(teste2, columns=['nick'])
    tst2 = pd.DataFrame(tester2, columns=['classificacao'])
    ts = pd.concat([tst, tst2], axis=1, join_axes=[tst.index])
    result = [lol, ts]
    lol = pd.concat(result)
```

```python
lol.to_csv('baseDeNicks.csv')
```
